```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**XỬ LÝ BẢNG APPLICATION-BANKING**

```python
# đọc file Applications-Banking
df_application = pd.read_excel(r'D:\mindxdata\banking application\Applications-Banking.xlsx')
df_application
```

```python
#tìm missing
df_application.info()
```

=> DataFrame không có missing

```python
df_application.describe()
```

```python
#xóa cột không cần thiết trong bảng application (Action, Collateral Start Date, Conversion Factor,Covenant Type,Limits Rating, Rating,Expected Loss,Covenants,Application Error)
columns = ['Action', 'Collateral Start Date', 'Conversion Factor','Covenant Type','Limits Rating','Expected Loss','Covenants','Application Error','Margin Agreed']
df_application.drop(columns, inplace =True, axis=1)
```

```python
# group by Product ID 
Collateral = df_application.groupby('Product ID', as_index= False).agg(total_face_value = ('Face Value','sum'))
Collateral= Collateral.sort_values(by = ['total_face_value'],ascending=False)
Collateral= Collateral.reset_index(drop=True)
```

```python
#Vẽ sơ đồ kiểm định tính giá trị của tài sản thế chấp của sản phẩm
plt.figure(figsize=(21,10))

xs = Collateral['Product ID']
ys = Collateral['total_face_value']

plt.bar(xs,ys,width=0.5,color = 'blue')
plt.xlabel("Collateral Type")
plt.ylabel("total_face_value")
plt.grid()
plt.show()
```

```python
#xuất ra file csv đặt tên thành application1
df_application.to_csv("D:\\mindxdata\\application_banking.csv")
```
